```python
# creating new object x and y where x is to train our model and y is for prediction.
x = df.iloc[:,1:].values
y = df.iloc[:,0].values
```

```python
x
```

```python
y
```

# Splitting dataset into training set and test set

```python
# we are now importing train_test_split library to divide our dataset for training and testing

from sklearn.model_selection import train_test_split
```

```python
# we are going to split our dataset into two different sets that is traning dataset and test dataset.

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
```

```python
x_train
```

# Performing Fature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()

x_train[:,:12] = sc.fit_transform(x_train[:,:12])
x_test[:,:12] = sc.transform(x_test[:,:12])
```

# Deploying Model

```python
# we  are going to traing our dataset in multi linear regression model

from sklearn.linear_model import LinearRegression
```

```python
# we are calling the model
reg = LinearRegression()
```

```python
# We are traing our model 
reg.fit(x_train, y_train)
```

```python
# after traing we are going to test our model on x_test
y_pred = reg.predict(x_test)
```

```python
y_pred
```

```python
# We are compairing the predicted result (after traing our model) & the actual result.
# This shows the accuracy level our model that is 56%.
reg.score(x_test, y_test)
```
